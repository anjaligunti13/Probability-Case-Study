# marginal probability (single column)


```python
#1what is probabilites of getting rain tommorow?
sum(df['Outlook']=='Rain')/len(df)
```

```python
df['Outlook'].value_counts(normalize=True).loc["Rain"] #normalize to know th prob
```

```python
#2what is the probability of playimg cricket tommorow?
df["Play"].value_counts(normalize=True).loc["Yes"]
```

```python
#What is the probalilty that temparature is not cool tommorow?
sum(df['Temperature']!='Cool')/len(df)
```

```python
1-df['Temperature'].value_counts(normalize=True).loc['Cool']
```

# joint prob

```python
#1what is the probbility of playing cricket & temperature is hot?
sum((df["Play"]=="Yes")&(df['Temperature']=='Hot'))/len(df) #len(df)- probability
```

```python
pd.crosstab(index=df['Play'],columns=df['Temperature'],normalize=True).loc['Yes','Hot']
```

```python
#what is the probability of humidity is normal and temperature is cool?
sum((df["Humidity"]=="Normal")&(df['Temperature']=='Cool'))/len(df)
```

```python
pd.crosstab(index=df["Humidity"],columns=df['Temperature'],normalize=True).loc['Normal','Cool']
```

# conditional probability

```python
#what is the prob of playing cricket if it is raining?
pd.crosstab(index=df["Play"],columns=df["Outlook"],normalize='columns').loc['Yes',"Rain"]
#normalize=coloums becoz it is condition already happening
```

```python
#what is the probability of not playing cricket when humidity is high?
pd.crosstab(index=df['Humidity'],columns=df["Play"],normalize='index').loc["High","No"]
```
